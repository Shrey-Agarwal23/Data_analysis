# What is the probablity that the books with count of rating_5 greater than 40,000 have high average rating(i.e.greater than 4) ?

```python
count_avg_rat=books['average_rating'].loc[(books['average_rating']>=4)]
count_rating=books['ratings_5'].loc[(books['ratings_5']>=40000)]
prob=len(count_rating)/len(count_avg_rat)
print("Probablity is : ",prob)
```

# In the given  dataset, how many books of particular authors is given?

```python
authors=["George Orwell","Veronica Roth","Dan Brown","Charles Dickens","Cassandra Clare"]
author_1=books['authors'].loc[(books['authors']==authors[0])]
author_2=books['authors'].loc[(books['authors']==authors[1])]
author_3=books['authors'].loc[(books['authors']==authors[2])]
author_4=books['authors'].loc[(books['authors']==authors[3])]
author_5=books['authors'].loc[(books['authors']==authors[4])]
count=[len(author_1),len(author_2),len(author_3),len(author_4),len(author_5)]
print("Number of books by George Orwell in the given dataset is :",len(author_1))
print("Number of books by Veronica Roth in the given dataset is :",len(author_2))
print("Number of books by Dan Brown in the given dataset is :",len(author_3))
print("Number of books by Charles Dickens in the given dataset is :",len(author_4))
print("Number of books by Cassandra Clare in the given dataset is :",len(author_5))

plt.bar(authors,count)
plt.xticks(fontsize=8)
plt.xlabel('Authors')
plt.ylabel('Number of books')
```
